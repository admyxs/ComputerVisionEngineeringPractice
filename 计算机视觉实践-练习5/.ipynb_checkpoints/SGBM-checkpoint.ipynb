```python
import cv2
import numpy as np

def compute_disparity_sgbm(left_image, right_image):
    """
    使用SGBM算法计算视差图。

    参数:
    - left_image: 左眼图像，np.ndarray类型
    - right_image: 右眼图像，与left_image相同尺寸和类型的np.ndarray

    返回:
    - disparity_map: 视差图，np.ndarray类型
    """

    # 确保输入为灰度图像
    if len(left_image.shape) != 2 or len(right_image.shape) != 2:
        # 如果是彩色图像，则转换为灰度图
        left_gray = cv2.cvtColor(left_image, cv2.COLOR_BGR2GRAY)
        right_gray = cv2.cvtColor(right_image, cv2.COLOR_BGR2GRAY)
    else:
        left_gray, right_gray = left_image, right_image

    # 设置SGBM算法的参数
    window_size = 3
    min_disp = 0
    max_disp = 16 * 8  # 视差范围通常乘以16
    num_disp = max_disp - min_disp
    block_size = 3
    uniqueness_ratio = 15
    speckle_window_size = 0
    speckle_range = 2
    disp12_max_diff = 1

    # 创建SGBM对象并配置参数
    stereo = cv2.StereoSGBM_create(
        minDisparity=min_disp,
        numDisparities=num_disp,
        blockSize=block_size,
        P1=8 * 3 * window_size ** 2,
        P2=32 * 3 * window_size ** 2,
        disp12MaxDiff=disp12_max_diff,
        uniquenessRatio=uniqueness_ratio,
        speckleWindowSize=speckle_window_size,
        speckleRange=speckle_range,
        preFilterCap=63,
        mode=cv2.STEREO_SGBM_MODE_SGBM_3WAY
    )

    # 计算左右图像的视差
    disparity_map = stereo.compute(left_gray, right_gray).astype(np.float32) / 16.0

    return disparity_map
```

```python
# 示例代码：加载立体图像对，计算视差，并显示结果
if __name__ == "__main__":
    left_image_path = '3.png'
    right_image_path = '7.png'
    left_image = cv2.imread(left_image_path)
    right_image = cv2.imread(right_image_path)

    disparity_map = compute_disparity_sgbm(left_image, right_image)

    # 显示视差图并保存
    cv2.imshow('Disparity Map', disparity_map / disparity_map.max())
    cv2.imwrite('result.png', disparity_map)
```

```python
import matplotlib.pyplot as plt
from matplotlib.image import imread
m1=imread('3.png')
m2=imread('7.png')
m3=imread('result.png')

plt.figure(figsize=(8, 8))
plt.subplot(221)
plt.imshow(m1)
plt.title('Left Image')
plt.subplot(222)
plt.imshow(m2)
plt.title('Right Image')
plt.subplot(212)
plt.imshow(m3, cmap='gray')
plt.title('Disparity Map')
plt.show()
```
